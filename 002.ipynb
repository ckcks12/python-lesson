네 그리고 우리가 사용하는 csv 라이브러리에서도 딕셔너리로 불러와 보도록 하죠.

````python
data = []
with open('/Users/danny/Downloads/MSFT.csv') as file:
    reader = csv.DictReader(file)
    for r in reader:
        data.append(r)

print(data[0]['Date'])
print(data[1]['Open'])
print(data[1]['High'])
print(data[1]['Low'])
````

X 축과 Y 축이 필요해요.

X 축은 시간

Y 축은 가격 이 될거에요.

````python
X = []
Y = []
for d in data:
    X.append(d['Date'])
    Y.append(d['Close'])
print(X[:10])
print(Y[:10])
````

준비된 X Y 데이터로 이제 그래프를 그려볼까요?

이번엔 상한가/하한가도 그려볼게요.

X 축은 똑같으니 그대로 쓰고 Y 축만 상한가, 하한가를 각각 만들어주면 된답니다.

````python
high = []
low = []
for d in data:
    high.append(float(d['High']))
    low.append(float(d['Low']))
print(high[:10])
print(low[:10])
````

이제 아까 쓰던 차트에 이어서 추가해보도록 하죠.

음 너무 번잡하네요?

등락폭이 좀 잠잠해보이는 2019.09~2019.10 기간을 뽑아볼까요?

X Y 축을 새로 만들어야겠네요~

````python
X = []
Y = []
high = []
low = []
for d in data:
    if '2019-09-01' <= d['Date'] <= '2019-10-31':
        X.append(d['Date'])
        Y.append(float(d['Close']))
        high.append(float(d['High']))
        low.append(float(d['Low']))
print(X[:3])
print(Y[:3])
print(high[:3])
print(low[:3])
````

잘 나왔나요?

이제 그대로 다시 차트를 생성해서 그려볼게요~


````python
import pygal
chart = pygal.Line()
chart.title = '주식비서'
chart.x_labels = X
chart.add('prices', Y)
chart.add('상한가', high)
chart.add('하한가', low)
chart.render_in_browser()
````

저번에 했던 계산 프로그램이 실력으로 한번 최대 수익을 계산해볼까요?

위 기간동안 매일 매일 최저가에 사서 최고가에 팔았다고 생각하는 거에요.

````python
account = 1000
print('현재 가지고 있는 돈', account, '달러')
for i in range(len(high)):
    stock = account / low[i] # 최저가로 최대한 많이 사요
    account = account % low[i] # 살 수 없을 만큼 사고 나머지를 남겨요
    account += stock * high[i] # 최고가로 되팔아요
print('현재 가지고 있는 돈', account, '달러')
````
